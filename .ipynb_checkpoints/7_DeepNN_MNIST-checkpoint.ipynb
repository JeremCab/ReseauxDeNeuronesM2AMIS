## 2. Dataloaders

A **dataloader** creates batches of samples from a dataset so that they can be passed into a model.
- Create a train and test dataloaders using the following commands:
```
train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_dataloader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=True)
```
- Note that dataloaders are not subscriptable.
- Try to catch one batch of the dataloader and examine it.
- Write a function that reshapes a batch of size $64 \times 1 \times 28 \times 28$ into a tensor of size $64 \times 784$.<br>
(use `torch.squeeze()`, `torch.reshape()`, `torch.flatten()`, `torch.transpose()`, etc.)

````python
train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_dataloader = torch.utils.data.DataLoader(test_dataset, batch_size=64, shuffle=True)
````

````python
def reshape_batch(batch):
    
    # note that this function moodifies the batch in-place (no return)
    batch_size = batch[0].shape[0]
    batch[0] = batch[0].view(batch_size, -1)
````

## 3. Model

- Define and instanciate a **neural network** with the following feaures:
    - hidden layer 1: 256 neurons, relu activation function
    - hidden layer 2: 256 neurons, relu activation function
    - hidden layer 3: 128 neurons, relu activation function
    - hidden layer 4: 64 neurons, relu activation function
    - output layer: 10 neurons

````python
class Network(nn.Module):

    def __init__(self):
        
        super(Network, self).__init__()
        
        self.fc1 = nn.Linear(784, 256)
        self.fc2 = nn.Linear(256, 256)
        self.fc3 = nn.Linear(256, 128)
        self.fc4 = nn.Linear(128, 64)
        self.fc5 = nn.Linear(64, 10)

    def forward(self, x):
        
        x = self.fc1(x)
        x = nn.ReLU()(x)
        
        x = self.fc2(x)
        x = nn.ReLU()(x)
        
        x = self.fc3(x)
        x = nn.ReLU()(x)
        
        x = self.fc4(x)
        x = nn.ReLU()(x)
        
        x = self.fc5(x)
        
        return x
````

````python
network = Network()
````

````python
# Put network to GPU if exists
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
network.to(device)
````

## 6. Training

- Define a **training loop** for you model.
- The key instructions are:
    - ``optimizer.zero_grad()``: reset gradients to 0
    - ``loss.backward()``      : compute new gradients
    - ``optimizer.step()``     : uptdate weights
- Train the model for 20 epochs.
- Plot the training loss.

````python
def train(model, loss=None, optimizer=None, dataloader=None, nb_epochs=20):
    """Training loop"""
    
    loss_l = []
        
    for e in tqdm(range(nb_epochs)):
    
        for i, batch in enumerate(dataloader):   # iterate over dataloader

            reshape_batch(batch)
            samples, labels = batch[0], batch[1]

            outputs = model(samples)

            current_loss = loss(outputs, labels) #Â compute loss
            optimizer.zero_grad()                # reset gradients to 0
            current_loss.backward()              # computer gradients
            optimizer.step()                     # update weights

            if i % 100 == 0: #Â every 100 batches

                loss_l.append(current_loss.item())

                print(f"Epoch: {e}/{nb_epochs}. \t step: {i} \t loss: {current_loss.item()}")
            
    return loss_l
````

````python
loss_l = train(network, 
               loss=loss, 
               optimizer=optimizer, 
               dataloader=train_dataloader, 
               nb_epochs=20)
````

````python
# plot loss
plt.plot(range(len(loss_l)), loss_l, 
         linewidth=2.0, 
         label='training loss')

plt.xlabel("Epochs)", size=12)
plt.ylabel("Loss", size=12)
plt.legend()

plt.show()
````

## 6. Predictions

- Write a ``predict`` function
    - use ``with torch.no_grad():``
- Compute the accuracy and classification report of the train and test sets.

````python
def predict(model, dataloader):

    labels_l = []
    preds_l = []

    with torch.no_grad():
        
        for i, batch in enumerate(dataloader):   # iterate over dataloader

            reshape_batch(batch)
            samples, labels = batch[0], batch[1]
            
            labels_l.extend(labels.tolist())
            
            preds = model(samples)
            preds = torch.argmax(preds, dim=1)
            preds_l.extend(preds.tolist())
    
    return labels_l, preds_l
````

````python
train_labels_l, train_preds_l = predict(network, train_dataloader)
test_labels_l, test_preds_l = predict(network, test_dataloader)
````

````python
print(classification_report(train_labels_l, train_preds_l))
````

````python
print(classification_report(test_labels_l, test_preds_l))
````
