# Denoising Autoencoders (DAE)

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms

import matplotlib.pyplot as plt
```

## Load MNIST data

```python
transform = transforms.ToTensor()

#transform = transforms.Compose( [transforms.ToTensor(),
#                                 transforms.Normalize((0.5),(0.5))] )

mnist_data = datasets.MNIST(root='./data', train=True,
                            download=True,
                            transform=transform)
```

```python
len(mnist_data)
```

```python
train_set, val_set = torch.utils.data.random_split(mnist_data, [50000, 10000])

train_loader = torch.utils.data.DataLoader(dataset=train_set, batch_size=64, shuffle=True)
val_loader = torch.utils.data.DataLoader(dataset=val_set, batch_size=64, shuffle=True)
```

```python
data_iter = iter(train_loader)
images, labels = next(data_iter)
```

```python
images.shape, labels.shape
```

```python
torch.min(images), torch.max(images)
```

## Adding noise to data

```python
def add_noise(images):
    
    return images + 0.5 * torch.randn(*images.shape)
```

```python
for batch in train_loader:
    break

images = batch[0][:32].squeeze(dim=1) # half of batch (32 items)
images_noisy = add_noise(images).squeeze(dim=1)

images.shape, images_noisy.shape
```

```python
fig, axs = plt.subplots(8, 8, 
                        figsize=(8, 8), 
                        layout="constrained",
                        gridspec_kw={'wspace': 0.1, 'hspace': 0.1})

for i in range(8):
    
    if i % 2 == 0:
        i2 = i // 2
        samples = images[8*i2:8*i2+8, :, :]
    else:
        i2 = i // 2
        samples = images_noisy[8*i2:8*i2+8, :, :]
    
    for j in range(8):
    
        axs[i, j].imshow(samples[j, :, :])
        axs[i, j].set_xticks([])
        axs[i, j].set_yticks([])

fig.savefig('figures/DAE_noisy_data.pdf', format='pdf', bbox_inches='tight')
# plt.show()
```

## Model: DAE with convolutional layers

### Model

```python
class DenoisingAutoencoder_CNN(nn.Module):
    """Implements a CNN denoising autoencoder"""

    def __init__(self):
        """constructor"""

        super().__init__()

        # N, 1, 28, 28
        self.encoder = nn.Sequential(
            # -> N, 16, 14, 14
            nn.Conv2d(1, 16, 3, stride=2, padding=1),
            nn.ReLU(),
            # -> N, 32, 7, 7
            nn.Conv2d(16, 32, 3, stride=2, padding=1),
            nn.ReLU(),
            # -> N, 64, 1, 1
            nn.Conv2d(32, 64, 7)
            )

        # N , 64, 1, 1
        self.decoder = nn.Sequential(
            # -> N, 32, 7, 7
            nn.ConvTranspose2d(64, 32, 7),
            nn.ReLU(),
            # N, 16, 14, 14 (N, 16, 13, 13 without output_padding)
            nn.ConvTranspose2d(32, 16, 3, 
                               stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            # N, 1, 28, 28 (N,1, 27, 27)
            nn.ConvTranspose2d(16, 1, 3, 
                               stride=2, padding=1, output_padding=1),
            nn.Sigmoid()
            )

    def forward(self, x):
        """forward function"""

        encoded_data = self.encoder(x)
        decoded_data = self.decoder(encoded_data)

        return decoded_data


# Note: nn.MaxPool2d -> use nn.MaxUnpool2d, or use different kernelsize, stride etc to compensate...
# Input [-1, +1] -> use nn.Tanh
```

```python
model = DenoisingAutoencoder_CNN()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), 
                             lr=1e-3, 
                             weight_decay=1e-5)
```

### Training

- The dataset consists of pairs of clean and noisy data $(\mathbf{x}, \mathbf{\tilde x})$.
- The MSE loss between those in minimized.

```python
num_epochs = 20
outputs_l = []

for epoch in range(num_epochs):

    train_loss = []
    val_loss = []

    for (inputs, _) in train_loader:      # clean inputs           (1)

        inputs_noisy = add_noise(inputs)  # noisy inputs
        outputs = model(inputs_noisy)     # images of noisy inputs (2)
        
        # loss between images of noisy inputs and clean inputs
        loss = criterion(outputs, inputs) # loss between (2) and (1)
        train_loss.append(loss.item())

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    train_loss = torch.mean(torch.tensor(train_loss)).item()

    for (inputs, _) in val_loader:

        inputs_noisy = add_noise(inputs)
        outputs = model(inputs_noisy)

        loss = criterion(outputs, inputs)
        val_loss.append(loss.item())

    val_loss = torch.mean(torch.tensor(val_loss)).item()

    print(f'Epoch:{epoch+1} \t Train loss: {train_loss:.4f} \t \
                             Validation loss: {val_loss:.4f}')

    outputs_l.append((epoch, inputs_noisy, outputs, train_loss, val_loss))
```

### Results

```python
plt.figure(figsize=(5, 3))

epochs = [x[0] for x in outputs_l]
train_loss = [x[3] for x in outputs_l]
val_loss = [x[4] for x in outputs_l]

plt.plot(epochs, train_loss, label="training loss")
plt.plot(epochs, val_loss, label="validation loss")
plt.xlabel("epochs")
plt.ylabel("MSE loss")
plt.legend()
plt.savefig('figures/DAE_cnn_loss_1.pdf', format='pdf', bbox_inches='tight')
#plt.show()
```

```python
fig, axs = plt.subplots(6, 10, 
                        figsize=(10, 7), 
                        layout="constrained",
                        gridspec_kw={'wspace': 0.1, 'hspace': 0.1})

for i, k in enumerate([0, num_epochs//2, num_epochs-1]):

    inputs = outputs_l[k][1].detach().numpy()
    outputs = outputs_l[k][2].detach().numpy()

    axs[2*i, 0].set_title(f"Epoch {k+1}")
    axs[2*i, 0].set_ylabel("Noisy inputs")

    for j, item in enumerate(inputs):

        if j >= 10: break
        axs[2*i, j].imshow(item[0])
        axs[2*i, j].set_xticks([])
        axs[2*i, j].set_yticks([])

    axs[2*i + 1, 0].set_ylabel("Outputs")
    for j, item in enumerate(outputs):

        if j >= 10: break
        axs[2*i + 1, j].imshow(item[0])
        axs[2*i + 1, j].set_xticks([])
        axs[2*i + 1, j].set_yticks([])

fig.savefig('figures/DAE_cnn_1.pdf', format='pdf', bbox_inches='tight')
# plt.show()
```

## Noise2Noise Paradigm

- The dataset consists of pairs of independently noisy versions $(\mathbf{\bar x}, \mathbf{\tilde x})$ of a same unknown data $\mathbf{x}$.
- The MSE loss between those in minimized.

```python
model = DenoisingAutoencoder_CNN()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), 
                             lr=1e-3, 
                             weight_decay=1e-5)
```

```python
num_epochs = 20
outputs_l = []

for epoch in range(num_epochs):

    train_loss = []
    val_loss = []

    for (inputs, _) in train_loader:
        
        inputs_noisy1 = add_noise(inputs) # noisy inputs!          (1)
        inputs_noisy2 = add_noise(inputs) # other noisy inputs     
        outputs = model(inputs_noisy2)    # images of noisy inputs (2)
        
        # loss between noisy inputs and images of other noisy inputs
        loss = criterion(outputs, inputs_noisy1) # loss beween (2) and (1)
        train_loss.append(loss.item())

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    train_loss = torch.mean(torch.tensor(train_loss)).item()

    for (inputs, _) in val_loader:

        inputs_noisy1 = add_noise(inputs) # noisy inputs!          (1)
        inputs_noisy2 = add_noise(inputs) # other noisy inputs     (2)
        outputs = model(inputs_noisy2)    # images of noisy inputs (3)

        loss = criterion(outputs, inputs_noisy1) # loss beween (3) and (1)
        val_loss.append(loss.item())

    val_loss = torch.mean(torch.tensor(val_loss)).item()

    print(f'Epoch:{epoch+1} \t Train loss: {train_loss:.4f} \t \
                             Validation loss: {val_loss:.4f}')

    outputs_l.append((epoch, inputs_noisy1, outputs, train_loss, val_loss))
```

### Results

```python
plt.figure(figsize=(5, 3))

epochs = [x[0] for x in outputs_l]
train_loss = [x[3] for x in outputs_l]
val_loss = [x[4] for x in outputs_l]

plt.plot(epochs, train_loss, label="training loss")
plt.plot(epochs, val_loss, label="validation loss")
plt.xlabel("epochs")
plt.ylabel("MSE loss")
plt.legend()
plt.savefig('figures/DAE_cnn_loss_2.pdf', format='pdf', bbox_inches='tight')
#plt.show()
```

```python
fig, axs = plt.subplots(6, 10, 
                        figsize=(10, 7), 
                        layout="constrained",
                        gridspec_kw={'wspace': 0.1, 'hspace': 0.1})

for i, k in enumerate([0, num_epochs//2, num_epochs-1]):

    inputs = outputs_l[k][1].detach().numpy()
    outputs = outputs_l[k][2].detach().numpy()

    axs[2*i, 0].set_title(f"Epoch {k+1}")
    axs[2*i, 0].set_ylabel("Noisy inputs")

    for j, item in enumerate(inputs):

        if j >= 10: break
        axs[2*i, j].imshow(item[0])
        axs[2*i, j].set_xticks([])
        axs[2*i, j].set_yticks([])

    axs[2*i + 1, 0].set_ylabel("Outputs")
    for j, item in enumerate(outputs):

        if j >= 10: break
        axs[2*i + 1, j].imshow(item[0])
        axs[2*i + 1, j].set_xticks([])
        axs[2*i + 1, j].set_yticks([])

fig.savefig('figures/DAE_cnn_2.pdf', format='pdf', bbox_inches='tight')
# plt.show()
```

**Conclusion:**
- The DAE can learn to denoise data using clean data.<br>
  In this case the dataset consists of pairs of clean and noisy data $(\mathbf{x}, \mathbf{\tilde x})$. <br>
  During training, the MSE loss between those in minimized.
- The DAE can also learn to denoise data using noisy data.<br>
  In this case the dataset consists of pairs of independent noisy versions $(\mathbf{\bar x}, \mathbf{\tilde x})$ of a same unknown data $\mathbf{x}$. <br>
  During training, the MSE loss between those in minimized.
